```python
# @title 기본 제목 텍스트1111
import os
import asyncio
import websockets
import json
import y1
import time  # Added missing import
import struct # Added missing import
import psutil # Added missing import

uri = "wss://m1-foo.onrender.com/ws_s"  # Replace with your websocket URI and use wss for secure connection

start_time = last_noti_time = time.time()
bin = no = None
run_time = cpu_time = cpu_usage = 0.0
stage = move = 0
try:
  async with websockets.connect(uri) as websocket:
    #print(f"Connected to {uri}")
    while True:
      try:
        cur_time = time.time()
        elapsed_time = cur_time - start_time
        if elapsed_time > 300:
          break
        if time.time() - last_noti_time > 10:
          await websocket.send(json.dumps({"type": "noti",
                                  "name": os.popen("cat /etc/hostname").read().strip(),
                                  "stage": stage, "move": move,
                                  "run_time": f"{run_time:.2f}",
                                  "cpu_usage": f"{cpu_usage:.2f}",
                                  "cpu_time": f"{cpu_time:.2f}",
                                  "uptime": os.popen("uptime -p").read().strip()}))
          last_noti_time = time.time()
        # Use a small timeout for receiving messages. If no message, TimeoutError is raised.
        message = await asyncio.wait_for(websocket.recv(), timeout=0.1) # Changed timeout from 0 to 0.1
        json_message = json.loads(message)
        #print("Received JSON message:", json_message)
        if 'req' in json_message and json_message['req'] == 'run':
          bin = bytes.fromhex(json_message['bin'])
          no = int(json_message['no'], 16)
      except asyncio.TimeoutError:
        # If a timeout occurs (no message received within 0.1 seconds), process bin if available
        if bin:
          process = psutil.Process(os.getpid())
          process.cpu_percent(interval=None)
          start_run_time = time.time()
          out, no, ret = y1.foo(bin, no, 0)
          run_time = time.time() - start_run_time
          cpu_usage = process.cpu_percent(interval=None)
          cpu_times = process.cpu_times()
          move += 1
          if ret == 0:
            _, id, _, _ = struct.unpack("II32sI", out)
            print(f"{id:08x}...{no:08x}, Elapsed time: {elapsed_time} seconds")
            await websocket.send (json.dumps({"result": "True", "bin": out.hex(), "no": f"{no:08x}"}))
            bin = None
            stage += 1
          else:
            #print(f"...{no:08x}")
            await websocket.send (json.dumps({"result": "False"}))
          no += 1
        else:
          # If no message and no bin to process, sleep briefly to prevent busy looping
          await asyncio.sleep(0.01) # Reduced sleep slightly
        continue
except Exception as e:
  print(f"An error occurred: {e}. Retrying in 5 seconds...")
  await asyncio.sleep(1) # Changed sleep from 5 to 1 for quicker retry in case of general error
print("텍스트1111 Connection closed.")
```
